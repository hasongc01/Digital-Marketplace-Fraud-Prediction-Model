# 3. Customer Life Cycle Management

- tenure: total duration a user or account remains active, measured as current date - start date

```python
# create copy from cleaned client_df
df = client_df.copy()

# create tenure features
## tenure in days 
df['tenure_days'] = (df['date_activ'].max() - df['date_activ']).dt.days # df['date_activ'].max is current date

## tenure in years
df['tenure_years'] = df['tenure_days']/365
```

```python
# save for customer_lifecycle table

customers_lifecycle = df.copy()
```
